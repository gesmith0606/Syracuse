```python
from urllib import request
snapchaturl = "https://www.gameinformer.com/legacyreviews"
response = request.urlopen(snapchaturl)
print(type(response))
html = response.read().decode('utf8')
type(html)
```

```python
from bs4 import BeautifulSoup
htmlsoup = BeautifulSoup(html, 'html.parser')
type(htmlsoup)
```

```python
print(htmlsoup)
```

```python
print(htmlsoup.prettify())
```

```python
anchors = htmlsoup.find_all('a')
type(anchors)
```

```python
len(anchors)
```

```python
print(anchors[:10])
```

```python
for a in htmlsoup.find_all('a'):
 print(a.text)
```

```python
for a in anchors:
 print(a.text)
```

```python
#Example 2

from urllib import request
snapchaturl = "https://www.nytimes.com/svc/collections/v1/publish/https://www.nytimes.com/section/world/rss.xml"
response = request.urlopen(snapchaturl)
print(type(response))
html = response.read().decode('utf8')
type(html)
```

```python
from bs4 import BeautifulSoup
htmla = BeautifulSoup(html, 'html.parser')
type(htmla)
```

```python
print(htmla)
```

```python
anchors = htmla.find_all('category')
type(anchors)
```

```python
len(anchors)
```

```python
for a in htmla.find_all('category'):
 print(a.text)
```
