```python
db = client.bball
db.list_collection_names()
```

```python
coll = db.bbcoll
docs = coll.find()
doclist = [tweet for tweet in docs]
len(doclist)
```

```python
def print_tweet_data(tweets):
    for tweet in tweets:
        if tweet['retweet_count']>5000:
           print('\nDate: ',tweet['created_at'])
           print('From: ',tweet['user']['name'])
           print('Message: ',tweet['text'])
           print('Retweet Count:',tweet['retweet_count'])
      
```

```python
from nltk.tokenize import word_tokenize
word_tokens = []
for sent in tweetlist:
    print(word_tokenize(sent))
    word_tokens.append(word_tokenize(sent))
```

```python
from nltk.tokenize import TweetTokenizer
tknzr = TweetTokenizer(strip_handles=True)
from nltk import FreqDist
import nltk

token_tweets = []

for tweet in tweetlist:
    #text = tweet["text"]
    tokens = tknzr.tokenize(tweet)
    token_tweets.append(tokens)

print(token_tweets[:20])
import itertools


```

```python
token1  = list(itertools.chain(*token_tweets))
print(token1[:50])

fdist1 = nltk.FreqDist(token1)

fdist1.most_common(10)
```
